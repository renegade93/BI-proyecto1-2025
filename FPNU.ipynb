### Tokenizar
Usaremos "sentence tokenization" y "word tokenize" para poder mantener la mayor coherencia y sacar el mayor valor de nuestros datos.

```python
def remove_special_characters(text, remove_digits=False):
    pattern = r'[^a-zA-Záéíóúñü\s]' if not remove_digits else r'[^a-zA-Záéíóúñü\s]'
    text = re.sub(pattern, '', text)
    return text

def tokenize_text(df):
    df['textos'] = df['textos'].astype(str)
    sentences = df['textos'].apply(nltk.sent_tokenize)
    word_tokens = sentences.apply(lambda x: [nltk.word_tokenize(sentence) for sentence in x])
    return word_tokens
```

```python

df['textos'] = df['textos'].str.lower()
df['textos'] = df['textos'].apply(lambda x: remove_special_characters(x))
df['tokens'] = tokenize_text(df)
```

## Stop words

```python
nltk.download('stopwords')
stop_words = nltk.corpus.stopwords.words('spanish')
def eliminar_stop_words(texto):
    return [palabra for frase in texto for palabra in frase if palabra not in stop_words]

df['tokens'] = df['tokens'].apply(eliminar_stop_words)
```

```python
df['tokens'].head()
```

## Lematizacion

```python
nlp = spacy.load("es_core_news_sm")
def lematizar(lista_oraciones):
    joined_sentences = [" ".join(lista_oraciones)] 
    docs = nlp.pipe(joined_sentences, batch_size=1000)
    return [token.lemma_ for doc in docs for token in doc]

df["lemmas"] = df["tokens"].apply(lematizar)
```
